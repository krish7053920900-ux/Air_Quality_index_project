```python
def calculate_sub_index(cp, breakpoints):
    for bp in breakpoints:
        if bp['low'] <= cp <= bp['high']:
            return ((bp['I_high'] - bp['I_low']) /
                    (bp['high'] - bp['low'])) * (cp - bp['low']) + bp['I_low']
    return np.nan
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load data again (safe)
df = pd.read_csv("AQI_Project_Data.csv")

# Clean column names
df.columns = df.columns.str.lower().str.strip()

# Auto-detect PM2.5 column
pm25_col = None
for col in df.columns:
    if 'pm25' in col or 'pm2.5' in col or 'pm_25' in col:
        pm25_col = col
        break

print("Detected PM2.5 column:", pm25_col)

# Rename to standard
df.rename(columns={pm25_col: 'pm2.5'}, inplace=True)

# KPI 13 calculation
df['pm2.5'] = pd.to_numeric(df['pm2.5'], errors='coerce').fillna(0)

pollutant_avg = df[['pm2.5']].mean()
pollutant_pct = (pollutant_avg / pollutant_avg.sum()) * 100

# Bar chart
pollutant_pct.plot(kind='bar')
plt.title("PM2.5 Contribution (%)")
plt.ylabel("Percentage")
plt.show()
```

```python
safe_limit = 15
above_safe = (df['pm2.5'] > safe_limit).mean() * 100

below_safe = 100 - above_safe

pd.Series(
    {'Above Safe Limit': above_safe, 'Within Safe Limit': below_safe}
).plot(kind='pie', autopct='%1.1f%%', startangle=90)

plt.title("Days Above WHO Safe PM2.5 Limit")
plt.ylabel("")
plt.show()
```
