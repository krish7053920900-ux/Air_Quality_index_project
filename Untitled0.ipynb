```python
def calculate_sub_index(cp, breakpoints):
    for bp in breakpoints:
        if bp['low'] <= cp <= bp['high']:
            return ((bp['I_high'] - bp['I_low']) /
                    (bp['high'] - bp['low'])) * (cp - bp['low']) + bp['I_low']
    return np.nan
```

```python
df['Month'] = df['Date'].dt.month
monthly_aqi = df.groupby('Month')['AQI'].mean()

monthly_aqi.plot(marker='o', title="Monthly AQI Trend")
plt.show()
```

```python
df['Year'] = df['Date'].dt.year
yearly_aqi = df.groupby('Year')['AQI'].mean()

yearly_aqi.pct_change() * 100
```

```python
aqi_series = df.set_index('Date')['AQI']

model = ARIMA(aqi_series, order=(1,1,1))
model_fit = model.fit()

forecast = model_fit.forecast(steps=30)
forecast
```

```python
def aqi_category(aqi):
    if aqi <= 50: return "Good"
    elif aqi <= 100: return "Moderate"
    elif aqi <= 200: return "Poor"
    elif aqi <= 300: return "Very Poor"
    else: return "Severe"

df['AQI_Category'] = df['AQI'].apply(aqi_category)
```
